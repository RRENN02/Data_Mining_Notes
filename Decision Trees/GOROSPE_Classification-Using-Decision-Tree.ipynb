```python
from sklearn import datasets
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

> Investigating the dataframe

```python
df = pd.DataFrame(data=bc.data, columns=bc.feature_names)
```

```python
# train-test split
from sklearn.model_selection import train_test_split

x = df[df.columns[0:-1]]
y = df[df.columns[-1]]

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
X_train[:5]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train[:5]
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
```

```python
from sklearn import tree
text_rep = tree.export_text(classifier)
print(text_rep)
```

```python
import matplotlib.pyplot as plt 
fig = plt.figure(figsize = (10, 8))
_ = tree.plot_tree(classifier, feature_names=bc.feature_names, 
                   class_names=bc.target_names, 
                   filled=True)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

y_pred = classifier.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print('Accuracy: ', accuracy_score(y_test, y_pred))
```

```python
def tree_depth_tree_hyp_tuning(d) :
    classifier = DecisionTreeClassifier(max_depth=d, criterion='entropy')
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    print(confusion_matrix(y_test, y_pred))
    print('Accuracy: ', accuracy_score(y_test, y_pred))
    return accuracy_score(y_test, y_pred)
```

```python
tree_result = pd.DataFrame({'D': np.arange(1, 10)})
```

```python
tree_result['Accuracy'] = tree_result['D'].apply(tree_depth_tree_hyp_tuning)
tree_result
```

> Identify the highest possible accuracy by using different splitting criteria (Information and Gini Index) and tree pruning. Show the results of the classification and visualize the accuracy vs hyperparameters for comparison. 

```python
# evaluate accuracy for different splitting criteria and pruning (ccp_alpha)
criteria = ['gini', 'entropy']
ccp_alphas = [0.0, 0.001, 0.01, 0.1]
results = []
for crit in criteria:
    for alpha in ccp_alphas:
        clf = DecisionTreeClassifier(criterion=crit, ccp_alpha=alpha, random_state=42)
        clf.fit(X_train, y_train)
        acc = accuracy_score(y_test, clf.predict(X_test))
        results.append({'criterion': crit, 'ccp_alpha': alpha, 'Accuracy': acc})
results_df = pd.DataFrame(results)

# identify best configuration
best = results_df.loc[results_df['Accuracy'].idxmax()]
print("Best params:", best.to_dict())
best_clf = DecisionTreeClassifier(
    criterion=best['criterion'],
    ccp_alpha=best['ccp_alpha'],
    random_state=42
)
best_clf.fit(X_train, y_train)
y_best = best_clf.predict(X_test)
print("Confusion Matrix:\n", confusion_matrix(y_test, y_best))
print("Classification Report:\n", classification_report(y_test, y_best))

# plot accuracy vs ccp_alpha for each criterion
sns.lineplot(data=results_df, x='ccp_alpha', y='Accuracy', hue='criterion', marker='o')
plt.title("Accuracy vs ccp_alpha by Splitting Criterion")
plt.xlabel("ccp_alpha")
plt.ylabel("Accuracy")
plt.show()
```

> Hyperparameter tuning using Optuna (Automatic Hyperparameter tuning)

Gini impurity index

```python
import optuna
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.tree import DecisionTreeClassifier



def objective(trial):
    criterion = trial.suggest_categorical("criterion", ["gini"])
    max_depth = trial.suggest_categorical("max_depth", [None, 3, 5, 7, 10])
    min_samples_leaf = trial.suggest_categorical("min_samples_leaf", [1, 2, 4])
    ccp_alpha = trial.suggest_categorical("ccp_alpha", [0.0, 0.001, 0.01, 0.1])

    model = DecisionTreeClassifier(
        criterion=criterion,
        max_depth=max_depth,
        min_samples_leaf=min_samples_leaf,
        ccp_alpha=ccp_alpha,
        random_state=42
    )
    scores = cross_val_score(model, X_train, y_train, cv=5, scoring="accuracy", n_jobs=-1)
    return scores.mean()

study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=100)

print("Best hyperparameters:", study.best_params)
print("Best CV accuracy: {:.4f}".format(study.best_value))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, GridSearchCV



# Parameter grid for splitting criteria and pruning (ccp_alpha)
param_grid = {
    'criterion': ['gini'],
    'max_depth': [3],
    'min_samples_leaf': [1],
    'ccp_alpha': [0.01]
}

# Grid search with 5â€‘fold CV
dt = DecisionTreeClassifier(random_state=42)
grid = GridSearchCV(dt, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid.fit(X_train, y_train)

print("Best params:", grid.best_params_)
print("Best CV accuracy: {:.4f}".format(grid.best_score_))

# Evaluate on heldâ€‘out test set
best_dt = grid.best_estimator_
test_acc = best_dt.score(X_test, y_test)
print("Test accuracy: {:.4f}".format(test_acc))
```

```python
from optuna.importance import get_param_importances

# Visualize Optuna optimization history for the Giniâ€criterion study
gini_study = study
trials_df = gini_study.trials_dataframe(attrs=('number', 'value'))
sns.lineplot(data=trials_df, x='number', y='value', marker='o')
plt.title('Optuna Optimization History (Gini)')
plt.xlabel('Trial')
plt.ylabel('Accuracy')
plt.show()

# Plot hyperparameter importance for the Gini study
importances = get_param_importances(gini_study)
imp_series = pd.Series(importances).sort_values(ascending=False)
imp_series.plot(kind='bar')
plt.title('Hyperparameter Importance (Gini)')
plt.ylabel('Importance')
plt.show()
```
